#คดีฆาตกรรมในวันที่ 15 มกราคม 2018

```python
#สืบหาฆาตรกร ในคดี ฆาตรกรรม ที่เกิดขึ้นในวันที่ 15 มกราคม 2018 ที่เมือง SQL City โดยตรวจสอบ ข้อมูลคดี (crime_scene_report.csv)
csr_df['date'] = csr_df['date'].astype(str)
#แสดงผลการใช้ โดยใช้ตัวนำเนินการ 'และ' ระหว่าง 2 Conditons
result = csr_df[(csr_df['date'].str.contains('20180115')) &
                (csr_df['type'] == 'murder') &
                (csr_df['city'] == 'SQL City')]
result
```

```python
#ภาพจากกล้องวงจรปิดแสดงให้เห็นว่ามีพยาน 2 คน โดยพยานคนแรกพักอาศัยที่ บ้านหลังท้ายสุด ของถนน 'Northwestern Dr'
max_number = per_df[per_df['address_street_name'] == 'Northwestern Dr']['address_number'].max()''
result = per_df[(per_df['address_street_name'] == 'Northwestern Dr') & (per_df['address_number'] == max_number)]
result
```

```python
#คำสั่ง Sprit Column หน้าและหลัง โดยแบ่งชื่อ
per_df[['first_name', 'last_name']] = per_df['name'].str.split(' ', n=1, expand=True)
per_df
```

```python
#พยานบุคคลที่ 2 ชื่อ Annabel ที่อาศัยอยู่บริเวณซักที่ ในถนน 'Franklin Ave'
per_df[(per_df['first_name'] == 'Annabel') & (per_df['address_street_name'] == 'Franklin Ave')]
```
