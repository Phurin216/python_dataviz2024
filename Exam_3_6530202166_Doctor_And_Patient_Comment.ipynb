#ilter เอาข้อมูล comments ที่มีจำนวนการตอบกลับมากที่สุด top 3

```python
doc_df_filtered = doc_df.sort_values(['จำนวนการตอบกลับ'],ascending=False)
doc_df_top3 = doc_df_filtered.head(3)
doc_df_top3
```

# filter เอาข้อมูล comments ที่ยาวที่สุด Top10

```python
doc_df['comment_length'] = doc_df['comments'].str.len()
doc_df_sorted = doc_df.sort_values('comment_length', ascending=False)
top_10_comments = doc_df_sorted.head(10)
top_10_comments
```
