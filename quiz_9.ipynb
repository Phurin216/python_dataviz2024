````
count = 0
stories_with_mom_or_dad = []
for i in range(folk_df.shape[0]):
    title = folk_df.iloc[i, 0]
    if 'แม่' in title or 'พ่อ' in title:
        count += 1
        stories_with_mom_or_dad.append(title)

print(f"มีนิทานที่มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องทั้งหมด {count} เรื่อง")
print("รายชื่อนิทาน:")
for story in stories_with_mom_or_dad:
    print(story)
````

````
count_life_stories = 0
count_local_stories = 0

for i in range(folk_df.shape[0]):
    title = folk_df.iloc[i, 0]
    story_type = folk_df.iloc[i, 1]
    if ('แม่' in title or 'พ่อ' in title):
        if story_type == 'นิทานชีวิต':
            count_life_stories += 1
        elif story_type == 'นิทานประจำถิ่น':
            count_local_stories += 1


if count_life_stories > count_local_stories:
    print("มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องของนิทานชีวิตมากกว่า")
elif count_local_stories > count_life_stories:
    print("มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องของนิทานประจำถิ่นมากกว่า")
else:
    print("มีคำว่า 'แม่' หรือ 'พ่อ' ในชื่อเรื่องของนิทานชีวิตและนิทานประจำถิ่นเท่ากัน")
````

````
creator_counts = {}
for i in range(folk_df.shape[0]):
    title = folk_df.iloc[i, 0]
    creator = folk_df.iloc[i, 2]
    if ('แม่' in title or 'พ่อ' in title):
        if creator not in creator_counts:
            creator_counts[creator] = 0
        creator_counts[creator] += 1

max_count = 0
most_prolific_creator = ""
for creator, count in creator_counts.items():
    if count > max_count:
        max_count = count
        most_prolific_creator = creator

print(f"คนแต่งที่ชอบแต่งนิทานที่มีชื่อเรื่องว่า 'แม่' หรือ 'พ่อ' มากที่สุดคือ: {most_prolific_creator}")
````
